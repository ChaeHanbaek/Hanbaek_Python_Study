라인편집기
i:라인 삽입, 행번호와 문자열을 입력하면 그행에 한라인 추가
d: 한 라인 삭제, 행번호를 입력하면 그행을 삭제
r: 한라인 변경, 행번호와 문자열을 입력하면 그행의 내용을 변경
p: 현재내용 출력. 현재문서의 모든 내용을 라인 번호와 함꼐 출력
l: 파일입력. 지정된 파일로부터 라인을 읽어 들임
s: 파일 출력 지정된파일로 편집내용을 저장
q: 편집기 종료

```python
class 라인편집기:
  str="---------라인편집기---------\ni:라인 삽입, 행번호와 문자열을 입력하면 그행에 한라인 추가(입력한 행번호가 현재 마지막 행번호보다 크면 맨뒤에 추가)\nd: 한 라인 삭제, 행번호를 입력하면 그행을 삭제\nr: 한라인 변경, 행번호와 문자열을 입력하면 그행의 내용을 변경\np: 현재내용 출력. 현재문서의 모든 내용을 라인 번호와 함꼐 출력\nl: 파일입력. 지정된 파일로부터 라인을 읽어 들임\ns: 파일 출력 지정된파일로 편집내용을 저장\nq: 편집기 종료\n------------------"
  def __init__(self):
    self.f_list=[] #self.를 안붙이면 인스턴스 생성해도 다른작업전까지는 안만들어진 상태가 됨
  def l(self,file_name):
    try:
      self.f_list=[] #열때마다 초기화를 안해주면 반복해서 열때 값이 중첩됨
      f = open("{}.txt".format(file_name), mode='r',encoding='utf-8-sig')
      for line in f:
        line=line.strip('\n') #줄바꿈제거
        self.f_list.append(line)
      print("{}.txt열기를 성공하였습니다.".format(file_name))
      print(self.f_list)
      return self.f_list
    except:
      print("{}.txt열기를 실패하였습니다.".format(file_name))

  def s(self,file_name):
    try:
      f = open("{}.txt".format(file_name),mode='w',encoding='utf-8-sig')
      for i in range(len(self.f_list)):
        if i<len(self.f_list)-1:
          f.write(str(self.f_list[i])+'\n')
        else:
          f.write(str(self.f_list[i]))
      print("{}.txt에 저장하는데 성공하였습니다.".format(file_name))
    except:
      print("{}.txt에 저장하는데 실패하였습니다.".format(file_name))

  def i(self,i,str):
    if i>0: #i는 1부터 시작하는 행의 번호이므로 인덱스로 쓰려면 -1로 보정해야함
      print("{}행에 {}값을 넣었습니다.".format(i,str))
      self.f_list.insert(i-1,str)
    else:
      print("입력값이 잘못되었습니다.")

  def d(self,i):
    if i>0 and i <= len(self.f_list): #지우는 작업은 리스트의 범위를 초과해서도 안되니 조건추가
      print("{}행의 값{}을 지웠습니다.".format(i,self.f_list[i-1]))
      del(self.f_list[i-1])
    else:
      print("입력값이 잘못되었습니다.")

  def r(self,i,str):
    if i>0 and i <= len(self.f_list):
      print("{}행의 값 {}을 {}로 바꿨습니다.".format(i,self.f_list[i-1],str))
      self.f_list[i-1]=str
  
  def p(self):
    for i in range(len(self.f_list)):
      print("{}행 : {}".format(i+1,self.f_list[i]))
  
  #비어있는지 확인용
  def e(self):
    t_list=self.f_list
    if len(t_list)==0:
      print("데이터가 들어있지 않습니다. 먼저 파일을 l로 불러오세요.")
      return False
    else:
      return True
```

```python
line = 라인편집기()
line.e()
a=line.l("abc")
line.p()
print(line.str)

line.i(3,343434)
line.p()

line.d(4)
line.p()

line.r(9,"구구구구")
line.p()

line.s("bcd")
```

```python
l=라인편집기()
while True:
  print(l.str)
  w=input("실행할 작업을 입력해주세요.")
  if w=='i':
      if l.e()==True:
        행=int(input("넣을 행번호를 입력해주세요. :"))
        값=input("추가할 문자열을 입력해주세요:")
        l.i(행,값)
        l.p()
      else:
        continue
  elif w=='d':
      if l.e()==True:
        행=int(input("지울 행번호를 입력해주세요. :"))
        l.d(행)
        l.p()
      else:
        continue
  elif w=='r':
      if l.e()==True:
        행=int(input("바꿀 행번호를 입력해주세요. :"))
        값=input("바꿀 문자열을 입력해주세요:")
        l.r(행,값)
        l.p()
      else:
        continue
  elif w=='p':
    l.p()
  elif w=='l':
    lname=input("열어볼 파일명을 입력해주세요.")
    l.l(lname)
    l.p()
  elif w=='s':
    sname=input("저장할 파일명을 입력해주세요.")
    l.s(sname)
  elif w=='q':
    print("편집기를 종료합니다.")
    break
  else:
    print("올바른 작업을 입력해주세요.")
```
