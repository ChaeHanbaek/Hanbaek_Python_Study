```python
#교수님 maxheap구현응용해 minheap 만들어야함
class MaxHeap:
  def __init__(self):
    self.heap=[]
    self.heap.append(0)

  def size(self) :
    return len(self.heap)-1
  def isEmpty(self):
    return self.size() ==0
  def Parent(self, i):
    return self.heap[i//2]
  def Left(self,i):
    return self.heap[i*2]
  def Right(self,i):
    return self.heap[i*2+1]
  def display(self,msg='힙 트리: '):
    print(msg, self.heap[1:])

  def insert(self,n):
    self.heap.append(n)
    i = self.size()
    while (i != 1 and n <= self.Parent(i)):
      self.heap[i]=self.Parent(i)
      i = i//2
    self.heap[i]=n

  def delete(self): #최상위 루트 노드 반환후 삭제
    parent = 1
    child =  2
    if not self.isEmpty():
      #루트와 마지막을 가지고옴
      hroot = self.heap[1]
      last= self.heap[self.size()]
      #이를 바탕으로 자식이 사이즈보다 커질 때까지(루트를 타고 쭉 내려감) 아래 연산 수행
      while (child <= self.size()):
        #왼쪽보다 오른쪽이 더클때 차일드값 +1(오른쪽 자식은 부모인덱스 *2 +1
        if child < self.size() and self.Left(parent)>=self.Right(parent):
          child +=  1
        #마지막 값이 자식값보다 크면 종료
        if last >= self.heap[child]:
          break;
        # parent와 차일드 값 변경          
        self.heap[parent] = self.heap[child]
        parent = child
        #왼쪽을 기준으로 내려감(왜냐하면 왼쪽자식은 부모인덱스*2이므로)
        child *=2; 
    self.heap[parent] = last
    self.heap.pop(-1)
    return hroot

    #위 연산을 그림을 그리면서 이해하는게 쉽다.
```

```python
#Minheap
class MinHeap:
  def __init__(self):
    self.heap=[]
    self.heap.append(0)

  def size(self) :
    return len(self.heap)-1
  def isEmpty(self):
    return self.size() ==0
  def Parent(self, i):
    return self.heap[i//2]
  def Left(self,i):
    return self.heap[i*2]
  def Right(self,i):
    return self.heap[i*2+1]
  def display(self,msg='힙 트리: '):
    print(msg, self.heap[1:])

  def insert(self,n):
    self.heap.append(n)
    i = self.size()
    while (i != 1 and n > self.Parent(i)):
      self.heap[i]=self.Parent(i)
      i = i//2
    self.heap[i]=n

  def delete(self): #최상위 루트 노드 반환후 삭제
    parent = 1
    child =  2
    if not self.isEmpty():
      #루트와 마지막을 가지고옴
      hroot = self.heap[1]
      last= self.heap[self.size()]
      #이를 바탕으로 자식이 사이즈보다 커질 때까지(루트를 타고 쭉 내려감) 아래 연산 수행
      while (child <= self.size()):
        #왼쪽보다 오른쪽이 더클때 차일드값 +1(오른쪽 자식은 부모인덱스 *2 +1
        if child < self.size() and self.Left(parent)<self.Right(parent):
          child +=  1
        #마지막 값이 자식값보다 크면 종료
        if last >= self.heap[child]:
          break;
        # parent와 차일드 값 변경          
        self.heap[parent] = self.heap[child]
        parent = child
        #왼쪽을 기준으로 내려감(왜냐하면 왼쪽자식은 부모인덱스*2이므로)
        child *=2; 
    self.heap[parent] = last
    self.heap.pop(-1)
    return hroot

    #위 연산을 그림을 그리면서 이해하는게 쉽다.
```

```python
#허프만 코드 by 힙
def make_tree(freq):
  heap = MinHeap()
  for n in freq:
    heap.insert(n)

  for i in range(0,n):
    e1 = heap.delete()
    e2 = heap.delete()
    heap.insert(e1+e2)
    print(" (%d+%d)" % (e1, e2))

#교수님이 준 코드는 01식으로 전환하는게 빠져있음, 이를 바탕으로 그거까지 구현해봐야함
```
