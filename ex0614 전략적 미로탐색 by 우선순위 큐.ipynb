```python
#우선순위 선형 큐
#미로 탐색을 위해 [x,y,d]값을 저장해서 사용할 예정
#d값이 낮을수록 우선순위 높음
class PriorityQueue:
  def __init__(self):
    self.items=[] #[x,y,d]저장
  
  def isEmpty(self):
    if len(self.items)==0:
      return True
    else:
      return False
  
  def size(self):
    return len(self.items)

  def clear(self):
    self.items=[]
  
  def enqueue(self,e):
    self.items.append(e)
  
  def findMaxPriority(self):
    if self.isEmpty():
      return None
    else:
      pindex=0
      for i in range(len(self.items)):
        if self.items[i][2] < self.items[pindex][2]: #d값이 작을수록 우선도 높음
          pindex=i
      return pindex

  def dequeue( self ):		
      highest = self.findMaxPriority()		
      if highest is not None :
          return self.items.pop(highest)	#[x,y,d]값을 반환하고 제거

  def peek( self ):				
      highest = self.findMaxPriority()	
      if highest is not None :
          return self.items[highest] #[x,y,d]값을 반환
```

```python
#0: 이동가능, 1:이동불가, 2: 시작위치, 9: 탈출위치
map=[[1,1,1,1,1,1],
     [2,0,0,0,0,1],
     [1,0,1,0,1,1],
     [1,1,1,0,0,9],
     [1,1,1,0,1,1],
     [1,1,1,1,1,1]]
maze_size=6
```

```python
def explore(맵,좌표): ##맵에서 해당좌표에서 어디로 갈수 있는지 여부 체크
  i=좌표[0] #행
  j=좌표[1] #열
  up=False
  down=False
  left=False
  right=False
  #상- 열 모서리:[0][j]
  if i-1>0:
    if 맵[i-1][j]!=1:
      up= True
    else:
      up= False
  elif i-1<0:
    up= False

  #하- 열 모서리:[len(map)][j]
  if i+1<len(맵): #len(맵)은 x의 행의 수를 나타냄
    if 맵[i+1][j]!=1:
      down=True
    else:
      down= False
  if i+1>=len(맵):
    down= False

  #좌- 행 모서리:[i][0]
  if j-1>0:
    if 맵[i][j-1]!=1:
      left= True
    else:
      left= False
  elif j-1<0:
    left= False
  
  #우- 모서리행 [i][len(맵[i])]
  if j+1<len(맵[i]):
    if 맵[i][j+1]!=1:
      right= True
    else:
      right= False
  elif j+1>=len(맵[i]):
    right= False
  return up, down, left, right

def move(좌표,방향,거리=1):
  #좌표=[행,열]
  #좌표[0]=행 좌표[1]=열
  if 방향=='up':
    좌표[0]-=거리
  elif 방향=='down':
    좌표[0]+=거리
  elif 방향=='left':
    좌표[1]-=거리
  elif 방향=='right':
    좌표[1]+=거리
  #print("{}로 이동합니다.".format(방향))
  return 좌표

def dspy3(맵,탐험됨,뷰):
  ss=' \| '
  for n in range(len(맵)):
    ss+=' '+str(n)+' '
  print(ss)
  print("---------------------")
  for i in range(len(맵)):
    t_lst=[] #열단위로 list에 넣어서 출력
    for j in range(len(맵[i])):
      t_lst.append('? ') #일단 물음표만들음
      #탐색 스택에 저장된 요소만 맵에서 읽어와서 대조
      for xx in 뷰: 
        if i==xx[0] and j==xx[1]:
          if 맵[i][j]==1:
            t_lst[j]='■ '
          elif 맵[i][j]==0:
            t_lst[j]='  '
          elif 맵[i][j]==9:
            t_lst[j]='x '      
          elif 맵[i][j]==2:
            t_lst[j]='e '           
      #탐험됨에 있는 좌표 1,2,3,4,5로 경로 표시하기
      TF=False
      for k in range(1,len(탐험됨)): #시작점 0은 딱히 기록안하고 넘어가기위해 1로 시작
        if i==탐험됨[k][0] and j==탐험됨[k][1]:
          if  k==len(탐험됨)-1:
            t_lst[j]='H '
          elif k<10:
            t_lst[j]='{} '.format(k)
          else:
            t_lst[j]='{}'.format(k)
          TF=True
    #list형식으로 출력하면 비율이 안좋음
    st=' {}| '.format(i) #스트링으로 바꿔서 출력
    for x in t_lst:
      st+=' '+x #스페이스 양쪽에 넣어야 넓이가 딱 적당함
    print(st)

def Echeck(좌표,탐험됨):
  if 좌표 in 탐험됨:
    return True
  else:
    return False

def 상하좌우(맵,좌표): ##지정된 좌표값 중심으로 상하좌우좌표값 list던져줌
  i=좌표[0] #행
  j=좌표[1] #열
  t_lst=[]
  if i-1>=0:
    t_lst.append([i-1,j])
  if i+1<len(map):
    t_lst.append([i+1,j])
  if j-1>=0:
    t_lst.append([i,j-1])
  if j+1<len(map):
    t_lst.append([i,j+1])
  return t_lst
```
