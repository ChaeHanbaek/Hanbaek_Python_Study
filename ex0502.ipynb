```python
#위 수식 함수화

def 랜덤함수 (_a,_b,_k): #_a~_b범위에서 중복없이 랜덤으로 _k개 뽑아 리스트로 돌려주는 함수
  import random
  _list=set()
  while len(_list) < _k:
    _list.add(random.randint(_a,_b)) 
  _list=list(_list) #set함수를 list형으로 반환\
  return _list
```

```python
import csv
r1=랜덤함수(1,10,5)
r2=랜덤함수(20,30,3)
r3=랜덤함수(1,100,10)
print(r1)
print(r2)
print(r3)

f=open("test.csv",mode="w",newline="",encoding="utf-8-sig")
wr=csv.writer(f)
wr.writerow(r1)
wr.writerow(r2)
wr.writerow(r3)

f.close()

dd=open("test.csv",mode="r",encoding="utf-8-sig")
rr=csv.reader(dd)
for line in rr:
  print(line)

dd.close()
```
