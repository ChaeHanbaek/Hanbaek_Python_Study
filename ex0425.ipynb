```python
#랜덤하게 숫자생성하고 업다운으로 숫자맞추는 게임 만들기
import random
num = random.randint(0,100) #랜덤모듈 가져옴, 거기에 내장된 함수 불러옴
print(num)
while(True):
  x=int(input("0~100까지 범위의 숫자를 맞춰보세요 -> ")) #input은 기본적으로 str으로 받아오기 때문에 int()로 강제 형변환 해줘야함
  if x == num:
    print("정답입니다.")
    break;
  elif x>num:
    print("정답보다 큽니다.")
  elif x<num:
    print("정답보다 작습니다.")
```

```python
def fooc():
  c=20

  def bar():
    global c
    c=25

  print("before calling bar: ",c)
  print("calling bar now")
  bar()
  print("after calling bar: ",c)

fooc()
print("c in main: ",c)
```

```python
def outer():
  d="local"
  def inner():
    nonlocal d
    d ="nonlocal"
    print("inner:",d)
  
  print("before inner loading outer:",d)
  inner()
  print("after inner loading outer:",d)

outer()
```
