```python
#추상 자료형
class Bag:
  bag = [] #클래스 변수
  def __init__(self): #특수함수 - 클래스 인스턴스화시, 자동으로 호출되는 함수
    self.bag = [] # 인스턴스 변수, 자료구조
  def clear(self): #self의 용도 : 클래스내에서 클래스가 가지고 있는 다른 요소 호출하기 위해 사용,받아오는 인자로 self 지정안하면 self.를 못건드림
    self.bag=[] #이클래스에서 self.bag=Bag.bag self를 쓰는 이유는 클래스명을 바꿔도 전체를 다시 수정안해도 되서 그럼
  def insert(self,e):
    self.bag.append(e)
  def remove(self,e):
    _t=False
    if self.contains(e) == True:
      print("{}이/가 제거되었습니다.".format(e))
      self.bag.remove(e)
      _t=True
      return e
    if _t==False:
      print("{}이/가 없습니다.".format(e))
    
  def contains(self,e):
    """
    for x in self.bag:
      if x==e:
        return True
        break
    return False 를 아래와 같이도 표현가능
    """
    
    if e in self.bag: #이런식으로 if 문 사용 가능
      return True
    else:
      return False

  def count(self):
    _t = len(self.bag)
    return _t
```

```python
x = Bag()
print(x.bag)
a=["지갑","휴대폰","동전","껌"]
for n in a:
  x.insert(n)
print(x.bag)
x.remove("지갑")
x.remove("지갑")
print(x.bag)
print(x.count())
```

```python
class List:
  def __init__(self):
    self.list=[]
  
  def insert(self,pos,e):
    if pos <len(self.list):
      print("list {}번째에 {}를 넣었습니다.".format(pos,e))
      self.list.insert(pos,e)
    else:
      print("{}가 list의 크기보다 큽니다.".format(pos))
       
  def delete(self,pos):
    if pos < len(self.list):
      _temp=self.list[pos]
      print("{}를 삭제했습니다.".format(self.list[pos]))
      del(self.list[pos]) #pop(pos)로 del하고 return 대체가능
      return _temp

    else:
      print("{}가 list의 크기보다 큽니다.".format(pos))

  def isEmpty(self):
    if len(self.list)==0:
      print("리스트가 비어있습니다.")
      print(self.list)
    else:
      print("리스트가 비어있지 않습니다.")
      print(self.list)

  def getEntry(self,pos):
    if pos < len(self.list):
      print("{}번째 요소는 {}입니다.".format(pos,self.list[pos]))
      return self.list[pos]
    else:
      print("{}가 리스트의 크기보다 큽니다.".format(pos))

  def size(self):
    print("list의 크기는 {}입니다.".format(len(self.list)))
    return len(self.list)

  def clear(self):  
    print("list가 초기화되었습니다.")
    self.list=[]

  def find(self,item):
    for i in range(len(self.list)):
      if self.list[i]==item:
        print("{}은 {}번째 요소입니다.".format(item,i))
        return i
        break
      elif i==len(self.list)-1:
        print("{}가 없습니다.".format(item))

  def replace(self,pos,item):
    if pos < len(self.list):
      print("{}번째 요소 {}를 {}로 바꿨습니다.".format(pos,self.list[pos],item))
      self.list[pos]=item
    else:
      print("{}가 list의 크기보다 큽니다.".format(pos))

  def sort(self,mode="up"):
    if mode == "up":
      self.list.sort()
      print("list를 오름차순 정렬했습니다.")
      print(self.list)

    elif mode == "down":
      self.list.sort(reverse=True)
      print("list를 내림차순 정렬했습니다.")
      print(self.list)

    else:
      print("mode변수가 잘못지정되었습니다. 'up'혹은 'down'이어야 합니다.")

  def merge(self,lst):
    print("{}을 list에 합쳤습니다.".format(lst))
    self.list.extend(lst)

  def display(self):
    print(self.list)

  def append(self,e):
    print("{}을 list 맨뒤에 넣었습니다.".format(e))
    self.list.append(e)
```

```python
L=List()
L.display()

L.insert(7,6)
print(L.list)
a=[1,2,3,5,4,3,2]
for i in range(len(a)):
  L.insert(i,a[i])

L.display()
print(L.delete(4))
L.display()
L.delete(4)

print(L.getEntry(4))
print(L.size())

L.find(4)
L.find(5)

L.replace(4,5)
L.display()

L.sort()
L.sort("down")


b=[7,8,9,10]
L.merge(b)

L.display()

L.append(11)
L.display()

L.isEmpty()
L.clear()
L.isEmpty()
```
