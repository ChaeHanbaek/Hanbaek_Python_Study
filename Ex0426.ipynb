```python
sheet = open("test.csv",mode='w',encoding='UTF-8-sig') #엑셀에서 csv파일이 안깨지게 하려면 단순 utf-8이 아니라 utf-8-sig로 해야한다.
sheet.write("이름,학번,성별\n")
sheet.write("채한백,2201340021,남\n") #csv는 comma으로 열 구분, tsv는 tab으로 열 구분
sheet.write("한백채,2021340021,남\n") #새로운 행을 추가하고싶으면 \n을 써야한다.

sheet.close()
"""최선의 해결책은 byte order mark (BOM)을 사용하는 것이다.
BOM은 문서 앞단에서 해당 문서는 한 글자가 몇 바이트로 작성되었는지를 명시해 준다. 
따라서 한 문자를 표현하는데 서로 다른 byte를 쓰는 인코딩들 사이에서 발생할 수 있는 문제를 해결해 준다. 
이를 지원하는 encoding uption이 바로 utf-8-sig이다. 
따라서 다음과 같은 옵션을 사용하면 상호간에 아무 문제가 발생하지 않을 수 있다. 
여기서 "sig"는 "signature의 준말이다."""
```

```python
import sys

randomList = ['a',0,2]

for entry in randomList:
  try:
    print("이번 엔트리는",entry)
    r=1/int(entry)
    break
  except:
    print("오!",sys.exc_info()[0],"문제 발생")
    print("다음 엔트리")
    print()
print("{}의 역수는 {}입니다.".format(entry,r))
```

```python
#예외처리 없이 2의 배수만 3개를 저장할 수 있는 코드 작성, 예외처리 추가
#가정설정문 assert로 만들기 
numlist=[]
count=0
while count<3:
  try:
    num = int(input("2의 배수인 숫자를 입력해주세요 {}/3 :".format(count)))
    assert num%2==0, "2의배수가 아님"
    numlist.append(num)
    count +=1
  except:
    print("2의 배수를 입력해주세요.")



print(numlist)
```

```python
#raise로 만들기
numlist=[]
count=0
while count<3:
  try:
    num = int(input("2의 배수인 숫자를 입력해주세요 {}/3 :".format(count)))
    if num%2!=0:
      raise
    else:
      numlist.append(num)
      count +=1
  except:
    print("2의 배수를 입력해주세요.")


print(numlist)
```

```python
#인스턴스화 하지않고 정의한뒤 함수 사용 가능
class clc:
  num1=10
  num2=20
  def __init__(self):
    pass

  def add(_num1, _num2):

    print(_num1+_num2)
  def sub(_num1, _num2):
    print(_num1-_num2)
  def mul(_num1, _num2):
    print(_num1*_num2)
  def div(_num1, _num2):
    print(_num1/_num2)


clc.add(1,2)
clc.sub(3,4)
clc.mul(5,6)
clc.div(7,8)
```

```python
#self 이해하기
class Foo:
  def func1():
    print("func1")
  
  def func2(self): #메서드의 self로 전달되는 것은 인스턴스 자체의 주소
    print("func2")

#클래스 메서드를 인스턴스를 생성한 뒤 호출
f=Foo()

f.func2() # func2 메서드의 첫 번째 인자는 self지만 호출할 때는 아무것도 전달하지 않는 이유는 첫 번째 인자인 self에 대한 값은 파이썬이 자동으로 넘겨주기 때문입니다.
#f.func1() #따라서 아무런 인자값이 없는 func1()에선 자동으로 넘겨받는 값때문에 오류발생

#클래스 메서드를 인스턴스없이 직접 호출

Foo.func1() #클래스에서부터 직접 가져올때는 func1() 오류발생 안함
#Foo.func2() #직접 가져올때는 자동으로 self값 못넘겨받아서 에러 발생
#메서드의 self로 전달되는 것은 인스턴스 자체의 주소이므로
#self값에 이미 생성된 인스턴스의 참조변수를 넘겨줘야 에러 발생 안함 
Foo.func2(f)
```

```python
#특수함수 __로 시작하는 클래스 함수
class PP:
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y

  def __str__(self): #참조변수를 출력했을때 주소값대신 나올 문자열을 설정하는 특수함수
    return "({0},{1})".format(self.x,self.y)

  def __add__(self,other): #같은 클래스의 참조변수끼리 더했을때 작동하는 방법을 만드는 특수함수
    x= self.x+other.x
    y= self.y+other.y
    return PP(x,y)
                            
p1 = PP(2,3)
print(p1)
p2 = PP(3,4)
print(p1+p2)
```
