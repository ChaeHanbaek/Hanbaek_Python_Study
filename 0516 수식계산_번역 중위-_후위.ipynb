수식의 후위 표기법 (Postfix Notation)

1.   1+2*3 -> 1,2,+,3,*
2.   8,2,/,3,-,3,2,*,+
*   ()로 연산 우선권 부여하려면 list에 넣고 순서 정하는게 좋아보임

1. 후위표기법으로 계산하는 함수만들기
*숫자하고 연산자를 스택에 넣고 빼서 계산한후 최종적으로 결과값이 저장되게 
2. 중위표기법을 후위표기법으로 번역하는 함수 만들기






```python
#자료구조의 스택의 특징을 담아 만든 클래스
class STACK():
  def __init__(self):
    self.lst=[]
  
  def isEmpty(self):
    if len(self.lst)==0:
      return True
    else:
      return False
  
  def push(self,e):
    self.lst.append(e)
    #print("STACK 맨위에 {}을 추가하였습니다.".format(e))
  
  def pop(self):
    t_v=self.lst.pop()
    return t_v
    #print("{}을 STACK 맨 위에서 꺼냈습니다.".format(t_v))
    

  def peek(self):
    #print("{}은 STACK 맨 위에 있는 요소입니다.".format(self.lst[(len(self.lst)-1)]))
    return self.lst[len(self.lst)-1]

  def size(self):
    t_count=0
    for x in self.lst:
      t_count +=1
    return t_count

  def clear(self):
    self.lst=[]

  def display(self):
    print(self.lst)
```

```python
#후위 표기식으로 써진 수식 리스트형식으로 받아서 계산하는 함수
def postfix_cal(clst):
  s=STACK() #스택
  for c in clst:
    
    if type(c)==int:
      s.push(c)
    elif c in '+-*/':
      b=s.pop()
      a=s.pop()
      print("{} {} {} 연산".format(a,c,b))
      t_results=0
      if c =='+':
        t_results=a + b       
      elif c=='-':
        t_results=a - b
      elif c=="*":
        t_results=a * b
      elif c=="/":
        t_results=a / b
      s.push(t_results)
      
    print(s.lst)
  return s.lst[0]
```

```python
ex1=[8,2,'/',3,'-',3,2,'*','+']
print(postfix_cal(ex1))
print(postfix_cal([1, 2, 3, '*', '+']))
```

```python
#중위표기법->후위표기법 번역기
def postfix_trans(lst):
  s=STACK()
  plst=[]
  count=0
  for c in lst:
    count +=1
    if len(s.lst)>0:
      LIFO=s.lst[len(s.lst)-1]
    else:
      LIFO='Nothing' #None으로 해두면 LIFO in '*/' 연산이 안됨
    #print(LIFO) #작동확인용

    if type(c)==int:
      plst.append(c)
    elif c in '+-': #*/우선권주기 : */가 마지막 자리에 저장되어있을때 +- 연산자 만나면 식에 들어가고 +-는 스택에 저장
      if LIFO in '*/':
        plst.append(s.pop())
      s.push(c)

    #중괄호, 대괄호도 해보기
    elif c in '({[*/':
      s.push(c)

    elif c in ')': #소괄호 우선권 주기 : '('는 스택에 저장한후 ')'만나면 저장되어있는 연산자 '('가 나올때까지 pop
      #'('가 pop로 튀어나올때까지 반복
      while True:
        tp=s.pop()
        if tp=='(':
          break
        else:
          plst.append(tp)

    #아직 중괄호, 대괄호 처리가 완벽하지 않음
    elif c in '}': #중괄호 우선권 주기 : '('는 스택에 저장한후 ')'만나면 저장되어있는 연산자 '('가 나올때까지 pop
      #'('가 pop로 튀어나올때까지 반복
      while True:
        tp=s.pop()
        if tp=='{':
          break
        else:
          plst.append(tp)
    
    elif c in ']': #대괄호 우선권 주기 : '('는 스택에 저장한후 ')'만나면 저장되어있는 연산자 '('가 나올때까지 pop
      #'('가 pop로 튀어나올때까지 반복
      while True:
        tp=s.pop()
        #print("확인: ",tp)
        #print("확인: ",s.lst) #확인 구문만 추가했는데 갑자기 정상작동?
        if tp=='[':
          break
        else:
          plst.append(tp)

    if count == len(lst): #마지막까지 갔을때 그때까지 저장되어있는 연산자들을 LIFO에따라 pop
      #print(s.lst) #작동확인용
      while True:
        plst.append(s.pop())
        if len(s.lst)==0:
          break

  #print(s.lst)  #작동확인용
  return plst
```

```python
중위1=[1,'+',2,'*',3]
중위2=[1,'*',2,'+',3]
중위3=['(',1,'+',2,')','*',3]
중위4=['(',1,'+',2,')','+',3,'*',4]
중위5=['(',1,'+',2,')','*',3,'*','(',4,'+',3,')']
print(postfix_trans(중위1))
print('------------------------')
print(postfix_trans(중위2))
print('------------------------')
print(postfix_trans(중위3))
print('------------------------')
print(postfix_cal(postfix_trans(중위3)))
print('------------------------')
print(postfix_trans(중위4))
print('------------------------')
print(postfix_cal(postfix_trans(중위4)))
print('------------------------')
print(postfix_trans(중위5))
print('------------------------')
print(postfix_cal(postfix_trans(중위5)))
```

```python
#아직 음수는 처리 못함, 음수를 (-5)식으로 넣는다는 약속을 가정하면 해결 가능할거 같음
def 스트링수식화(st):
  #일의 자리 숫자들은 list만해줘도 충분
  mlst=list(st)
  tlst=[]
  flst=[]
  for x in mlst:
    #print(x)
    if x.isdigit() ==False: #연산자나 괄호
      #넣기전에 tlst에 모인 숫자들 처리
      if len(tlst)>0: #tlst에 숫자 들어가있을때만 작동
        t=0
        for i in range(len(tlst)):
          j=len(tlst)-1-i
          t+= tlst[i]*(10**j)
        flst.append(t) #flst에 넣어줌
        tlst.clear() #초기화해야 제대로 작동함
      #연산자나 괄호 넣기
      flst.append(x)
    else:#숫자면
      tlst.append(int(x)) #일단 tlst에 모아두기
      #마지막 숫자면 연산자나 괄호 안만나도 tlst에 있는 수들 꺼내야함
      if x == mlst[len(mlst)-1]:
        t=0
        for i in range(len(tlst)):
          j=len(tlst)-1-i
          t+= tlst[i]*(10**j)
        flst.append(t) #flst에 넣어줌
        tlst.clear() #초기화해야 제대로 작동함
  #음수 '(','-','2',')'형태 찾아내서 -2 형태로 변환하기
  ty=0
  dlst=[]
  for i in range(len(flst)):
    if ty=="(" and flst[i]=='-':
      flst[i+1]=-flst[i+1]
      dlst.append(i) #부호'-'제거용
    ty=flst[i]

  for i in dlst:
    del flst[i],flst[i-1],flst[i] #각각 부호 -, 그앞 ( 그뒤 )제거

  return flst
```

```python
a="{(2+3)*4}+5*8"
print(스트링수식화(a))
b="{(2000+3)*40}+50*8"
print(스트링수식화(b))
c="(2+3)*{(4+5)*8}"
print(스트링수식화(c))
d="((-2)+3)*{(4+5)*8}"
print(스트링수식화(d))
e="[((-2)+3)*{(4+5)*8}]+{5*(4+3)}"
print(스트링수식화(e))
```
