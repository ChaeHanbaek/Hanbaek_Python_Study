```python
#preorder을 이용한 탐색 - 후위탐색
def search_value_bst(n,value):
  if n == None:
    return None
  elif value == n.value:
    return n

  res = search_value_bst(n.left, value)
  if res is not None:
    return res
  else:
    return search_value_bst(n.right, value) 
#순환구조
```

```python
#키를 이용한 탐색 - 순환
def search_bst(n,key):
  if n == None:
    return None
  elif key == n.key:
    return n
  elif key < n.key:
    return search_bst(n.left,key)
  else:
    return search_bst(n.right,key)

#키를 이용한 탐색 - 반복
def search_bst_iter(n,key):
  while n != None:
    if key == n.key:
      return n
    elif key < n.key:
      n= n.left
    else:
      n = n.right
  return None

#최대 키 노드 찾기 : 오른쪽(부모보다 큰쪽) 쭉 타고 감
def search_max_bst(n):
  while n != None and n.right != None:
    n = n.right
  return n

#최소 키 노드 찾기 : 왼쪽(부모보다 작은 쪽) 쭉 타고 감
def serach_min_bst(n):
  while n != None and n.left != None:
    n = n.left
  return n

#삽입 : 루트부터 시작해서 재귀함수를 통해 쭉 탐색해서 내려가 탐색에 실패한 위치(비어있는 자리)에 새로운 노드 삽입
def insert_bst(r,n): #r: 루트, n: 삽입할 값
  if n.key < r.key:
    if r.left is None:
      r.left = n
      return True
    else:
      return insert_bst(r.left,n) #재귀함수 

  elif n.key > r.key:
    if r.right is None:
      r.right = n
      return True
    else:
      return insert_bst(r.right, n) #재귀함수
  else:
    return False

#삭제
#단말 노드 삭제
def delete_bst_case1(parent, node, root):
  if parent is None:
    root = None
  else:
    if parent.left == node:
      parent.left = None
    else:
      parent.right = None
  return root

#비단말 노드(자식 1개)삭제
def delete_bst_case2(parent,node,root):
  if node.left is not None:
    child = node.left
  else:
    child = node.right

  if node == root:
    root = child
  else:
    if node is parent.left:
      parent.left = child
    else:
      parent.right = child

  return root

#비단말 노드(자식 2개) 삭제
def delete_bst_case3(parent,node,root):
  succp = node
  succ = node.right
  while (succ.left != None):
    succp = succ
    succ = succ.left
  if (succp.left == succ):
    succp.left = succ.right
  else:
    succp.right = succ.right
  
  node.key = succ.key
  node.value = succ.value
  node = succ;

  return root
# 1~3 통합
```
