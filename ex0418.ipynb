```python
#위에거 묶은 함수를 한번에 여러값 리턴하는 식으로 변형
def 평균(numbers):
  sum = 0
  for x in numbers:
    sum = sum+x
  average = sum/len(numbers)
  return average

def 분산(numbers):
  average = 평균(numbers)
  variance = 0
  for x in numbers:
    variance = variance+(x-average)*(x-average)
  variance = variance/len(numbers)
  return variance

def 표준편차(numbers):
  sd= 분산(numbers)
  sd = sd ** 0.5
  return sd

def 최댓값(numbers):
  max = numbers[0]
  for x in numbers:
    if x > max:
      max = x
  return max

def 최댓값인덱스(numbers):
  max = 최댓값(numbers)
  maxindex = 0
  for y in range(len(numbers)):
    if numbers[y]==max:
      maxindex = y
  return maxindex

def 최솟값(numbers):
  min = numbers[0]
  for x in numbers:
    if x < min:
      min = x
  return min

def 최솟값인덱스(numbers):
  min = 최솟값(numbers)
  minindex = 0
  for y in range(len(numbers)):
    if numbers[y]==min:
      minindex = y
  return minindex

#리스트 딥카피하기위한 copy 함수 추가, list 2= list1 식으로 복사하면 주소값 넘겨받는거라서 원본 list가 변화
def copy(numbers):
  copied = []
  for x in numbers:
    copied.append(x)
  return copied

def 오름차순정렬(numbers):
  upsort = copy(numbers)
  for x in range(len(upsort)):
    for i in range(len(upsort)-1):
      if upsort[i] > upsort[i+1]:
        n = upsort[i]
        upsort[i] = upsort[i+1]
        upsort[i+1] = n
  return upsort

def 내림차순정렬(numbers):
  downsort = copy(numbers)
  for x in range(len(downsort)):
    for i in range(len(downsort)-1):
      if downsort[i] < downsort[i+1]:
        n = downsort[i]
        downsort[i] = downsort[i+1]
        downsort[i+1] = n
  return downsort

#list 입력받고 출력부분 구현
numbers=[]
n=int(input("입력할 숫자의 갯수를 말해주세요: "))
print("도중에 중단하시려면 0 을 입력해주세요.")
for x in range(n):
  x=int(input())
  if x==0: #0들어오면 중단
    break
  else :
    numbers.append(x)
print(numbers)

# 1.평균 분산, 표준편차, 출력함수
def func01 (numbers):
  mean = 평균(numbers)
  val = 분산(numbers)
  sd = 표준편차(numbers)
  return mean, val, sd #여러 변수 값 return가능

f11, f12, f13 = func01(numbers) #이런 식으로 return값 받아옴, 받아올때 변수수가 적으면 그 뒤의 값은 못 받아옴
print("평균은 {}입니다.".format(f11))
print("분산은 {}입니다.".format(f12))
print("표준편차는 {}입니다.".format(f13))
print()

#2.최솟값, 최댓값, 최솟값인덱스, 최댓값인덱스 출력함수
def func2(numbers):
  max = 최댓값(numbers)
  maxindex = 최댓값인덱스(numbers)
  min = 최솟값(numbers)
  minindex = 최솟값인덱스(numbers)
  return [max,maxindex,min,minindex] #list로 묶어서 return 가능


maxmin=func2(numbers)
print("최댓값은 {}이고, 그 인덱스는 {}입니다.".format(maxmin[0],maxmin[1]))
print("최솟값은 {}이고, 그 인덱스는 {}입니다.".format(maxmin[2],maxmin[3]))
print()

#3.오름차순, 내림차순 정렬 출력함수
def func3(numbers):
  us = 오름차순정렬(numbers)
  ds = 내림차순정렬(numbers)
  return us, ds

f31, f32 = func3(numbers)
print("오름차순 정렬:",f31)
print("내림차순 정렬:",f32)
#오름차순 정렬, 내림차순 정렬은 참조변수로 원본 list를 건드리기때문에 실행된 순서에 따라 결과물이 바뀜
#각각 정렬된 list를 가지고 싶으면 copy 모듈을 import해서 사용하거나 그거랑 같은 함수를 만들어야됨
#copy함수를 만들어서 해결
```
