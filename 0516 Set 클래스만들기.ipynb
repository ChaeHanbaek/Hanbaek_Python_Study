```python
#교수님 요구 list를 통해 set성능 구현하기
class Set():
  def __init__(self):
    self.Set=[]
  def size(self):
    count=0
    for x in self.Set:
      count+=1
    return count
  def contain(self,e):
    return e in self.Set
  def insert(self,e):
    if self.contain(e) == True:
      pass
    else:
      self.Set.append(e)
  #setB와 같은 집합인지를 검사. 둘다 정렬한후, 모든값이 같으면 해결
  def equals(self,setB):
    #길이체크
    tf=True
    if len(self.Set) !=len(setB):
      tf=False
    #요소들이 들어있는지확인
    else:
      for x in self.Set:
        if x in setB:
          continue
        else:
          tf=False
    return tf
  #합집합
  def union(self,setB):
    uni=[]
    uni.extend(self.Set) #딥카피 방법
    for y in setB:
      if (y in uni) != True: #if문에 넣을때 ()안쳐주면 True, False 값이 아니라 이상하게 인식할 수 있음
        uni.append(y)
    return uni
      
  #교집합
  def intersect(self,setB):
    inter = []
    for x in self.Set:
      if (x in setB) == True:
        inter.append(x)
    return inter
  #차집합
  def difference(self,setB):
    diff=[]
    for x in self.Set:
      if (x in setB) ==False:
        diff.append(x)
    return diff
  def display(self):
    print(self.Set)

```

```python
setB=['a','b','c','d']
setA=Set()
print(setA.size())
XX=['a','c','e','f','g']
for x in XX:
  setA.insert(x)
print(setA.size())
setA.display()
print(setA.contain('e'))
print(setA.contain('d'))
print('--------------')
print(setA.equals(setB))
print(setA.equals(XX))
print(setA.Set)
print(setA.union(setB))
print(setA.intersect(setB))
print(setA.difference(setB))
setA.display()
XY=['c','a','g','f','e']
print(setA.equals(XY))
```
