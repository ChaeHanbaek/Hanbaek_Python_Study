```python
class Point:
  def __init__(self,x=0,y=0):
    self.x=x
    self.y=y
  def __str__(self):
    return "({0},{1})".format(self.x,self.y) #0,1로 안해두면 y값이 1씩 줄어듬

#Addition
  def __add__(self,other):
    x= self.x + other.x
    y= self.y + other.y
    return Point(x,y)
  
#Subtraction
  def __sub__(self,other):
    x= self.x - other.x
    y= self.y - other.y
    return Point(x,y)

#Multiplication
  def __mul__(self,other):
    x= self.x * other.x
    y= self.y * other.y
    return Point(x,y)

#Power    
  def __pow__(self,other):
    x= self.x ** other.x
    y= self.y ** other.y
    return Point(x,y)

#Division
  def __truediv__(self,other):
    x= self.x / other.x
    y= self.y / other.y
    return Point(x,y)

#Floor Division
  def __floordiv__(self,other):
    x= self.x // other.x
    y= self.y // other.y
    return Point(x,y)

#Remainder(modulo)
  def __mod__(self,other):
    x= self.x % other.x
    y= self.y % other.y
    return Point(x,y)

#Bitwise Left Shift
  def __lshift__(self,other):
    x= self.x << other.x
    y= self.y << other.y
    return Point(x,y)

#Bitwise Right Shift
  def __rshift__(self,other):
    x= self.x >> other.x
    y= self.y >> other.y
    return Point(x,y)

#Bitwise AND
  def __and__(self,other):
    x= self.x & other.x
    y= self.y & other.y
    return Point(x,y)

#Bitwise OR
  def __or__(self,other):
    x= self.x | other.x
    y= self.y | other.y
    return Point(x,y)

#Bitwise XOR
  def __xor__(self,other):
    x= self.x ^ other.x
    y= self.y ^ other.y
    return Point(x,y)

#Bitwise NOT
  def __invert__(self):
    x= ~self.x
    y= ~self.y
    return Point(x,y)

#Less than
  def __lt__(self,other):
    x= self.x < other.x
    y= self.y < other.y
    return Point(x,y)

#Less than or equal to
  def __le__(self,other):
    x= self.x <= other.x
    y= self.y <= other.y
    return Point(x,y)

#Equal to
  def __eq__(self,other):
    x= self.x == other.x
    y= self.y == other.y
    return Point(x,y)           

#Not equal to
  def __ne__(self,other):
    x= self.x != other.x
    y= self.y != other.y
    return Point(x,y)    

#Greater than
  def __gt__(self,other):
    x= self.x > other.x
    y= self.y > other.y
    return Point(x,y)         

#Greater than or equal to
  def __ge__(self,other):
    x= self.x >= other.x
    y= self.y >= other.y
    return Point(x,y)    
```

```python
p1=Point(1,2)
p2=Point(3,4)
print(p1)
print(p2)

print("__add__ +")
print(p1+p2)
print(p1.__add__(p2))

print("__sub__ -")
print(p1-p2)
print(p1.__sub__(p2))


print("__mul__ *")
print(p1*p2)
print(p1.__mul__(p2))

print("__pow__ **")
print(p1**p2)
print(p1.__pow__(p2))

print("__truediv__ ")
print(p1/p2)
print(p1.__truediv__(p2))

print("__floordiv__ //")
print(p1//p2)
print(p1.__floordiv__(p2))

print("__mod__ %")
print(p1%p2)
print(p1.__mod__(p2))

print("__lshift__ <<")
print(p1<<p2)
print(p1.__lshift__(p2))

print("__rshift__ >>")
print(p1>>p2)
print(p1.__rshift__(p2))

print("__and__ &")
print(p1&p2)
print(p1.__and__(p2))

print("__or__ |")
print(p1|p2)
print(p1.__or__(p2))

print("__xor__ ^")
print(p1^p2)
print(p1.__xor__(p2))

print("__invert__ ~")
print(~p1)
print(p1.__invert__())

print("__lt__ <")
print(p1<p2)
print(p1.__lt__(p2))

print("__le__ <=")
print(p1<=p2)
print(p1.__le__(p2))

print("__eq__ ==")
print(p1==p2)
print(p1.__eq__(p2))

print("__ne__ !=")
print(p1!=p2)
print(p1.__ne__(p2))

print("__gt__ >")
print(p1>p2)
print(p1.__gt__(p2))

print("__ge__ >=")
print(p1>=p2)
print(p1.__ge__(p2))
```
