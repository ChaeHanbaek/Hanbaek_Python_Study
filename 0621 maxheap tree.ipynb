힙을 저장하는 효과적인 구조는 배열
부모: 자식인덱스 /2
왼쪽 자식: 부모인덱스 *2
오른쪽 자식: 부모인덱스 *2 +1

```python
#교수님 maxheap구현
class MaxHeap:
  def __init__(self):
    self.heap=[]
    self.heap.append(0)

  def size(self) :
    return len(self.heap)-1
  def isEmpty(self):
    return self.size() ==0
  def Parent(self, i):
    return self.heap[i//2]
  def Left(self,i):
    return self.heap[i*2]
  def Right(self,i):
    return self.heap[i*2+1]
  def display(self,msg='힙 트리: '):
    print(msg, self.heap[1:])

  def insert(self,n):
    self.heap.append(n)
    i = self.size()
    while (i != 1 and n > self.Parent(i)):
      self.heap[i]=self.Parent(i)
      i = i//2
    self.heap[i]=n

  def delete(self): #최상위 루트 노드 반환후 삭제
    parent = 1
    child =  2
    if not self.isEmpty():
      #루트와 마지막을 가지고옴
      hroot = self.heap[1]
      last= self.heap[self.size()]
      #이를 바탕으로 자식이 사이즈보다 커질 때까지(루트를 타고 쭉 내려감) 아래 연산 수행
      while (child <= self.size()):
        #왼쪽보다 오른쪽이 더클때 차일드값 +1(오른쪽 자식은 부모인덱스 *2 +1
        
        
        )
        if child < self.size() and self.Left(parent)<self.Right(parent):
          child +=  1
        #마지막 값이 자식값보다 크면 종료
        if last >= self.heap[child]:
          break;
        # parent와 차일드 값 변경          
        self.heap[parent] = self.heap[child]
        parent = child
        #왼쪽을 기준으로 내려감(왜냐하면 왼쪽자식은 부모인덱스*2이므로)
        child *=2; 
    self.heap[parent] = last
    self.heap.pop(-1)
    return hroot

    #위 연산을 그림을 그리면서 이해하는게 쉽다.
```

```python
heap = MaxHeap()
data = [2,5,4,8,9,3,7,3]
print("[삽입 연산]: ",data)
for elem in data:
  heap.insert(elem)
heap.display('[삽입 후]: ')
heap.delete()
heap.display('[삭제 후]: ')
heap.delete()
heap.display('[삭제 후]: ')
```
