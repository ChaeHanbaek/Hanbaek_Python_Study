```python
class STACK():
  def __init__(self):
    self.list=[]
  
  def isEmpty(self):
    if len(self.list)==0:
      return True
    else:
      return False
  
  def push(self,e):
    self.list.append(e)
    print("STACK 맨위에 {}을 추가하였습니다.".format(e))
  
  def pop(self):
    nn=int(len(self.list)-1)
    t_v=self.list[nn]
    del self.list[nn]
    print("{}을 STACK 맨 위에서 꺼냈습니다.".format(t_v))
    return t_v

  def peek(self):
    print("{}은 STACK 맨 위에 있는 요소입니다.".format(self.list[(len(self.list)-1)]))
    return self.list[len(self.list)-1]

  def size(self):
    t_count=0
    for x in self.list:
      t_count +=1
    return t_count

  def clear(self):
    self.list=[]

  def display(self):
    print(self.list)
```

```python
SS=STACK()
print(SS.isEmpty())
SS.display()
X=['aaa','bbb','ddd','ccc']
for x in X:
  SS.push(x)
SS.display()
SS.pop()
SS.display()
SS.peek()
SS.display()
SS.clear()
SS.display()
```
