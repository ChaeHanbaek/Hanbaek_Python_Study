```python
class calc: #클래스 사용법 1. 함수처럼 사용, 2. 인스턴스화로 객체로 만들고, 객체내부에서 함수 사용
    #클래스 생성자
    def __init__(self):
        pass

    #클래스 자체함수라고 명시  
    def add1(self, num1, num2):
        res = num1+num2
        print("{} + {} = {}".format(num1, num2, res))
        return res

    def add2(num1, num2):
        res = num1+num2
        print("{} + {} = {}".format(num1, num2, res))
        return res

    def add3(num1, num2):
        res = calc.add2(num1, num2) #self가 아니라 클래스명.함수명으로 사용, 클래스내 위치를 알려주는것
        return res

    def add4(self, num1, num2):
        res = self.add1(num1,num2)
        return res
```

클래스를 함수처럼 사용

```python
print(calc.add2(1,2))
```

```python
print(calc.add1(None, 1, 2)) #self 들어가는 함수는 주소값에  None넣어서 사용 가능
```

```python
print(calc.add3(1, 2))
```

클래스를 인스턴스화시켜 사용할때는, self가 없는 함수는 사용 못함

```python
obj_calc = calc()
obj_calc.add1(1, 2)
```
