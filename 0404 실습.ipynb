```python
str='cold';

list_enumerate = list(enumerate(str)); #enumerate() 객체를 순차적으로 요소마다 인덱스를 부여해서 같이 저장
#바로 dic에 값넣을때 활용도 가능
print('list(enumerate(str) = ', list_enumerate);
print('len(str) =', len(str)); #len() 객체를 이루는 요소의 개수를 반환, 반복문 반복횟수나, 사전에 리스트 길이 지정할때 사용 가능
```
