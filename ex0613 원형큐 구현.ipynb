```python
class Queue:
  def __init__(self,n):
    self.Max_Qsize=n
    self.Queue=[None] * self.Max_Qsize
    self.front=0 #front는 맨 앞을 지울때 self.front=(self.front+1)%self.Max_Qsize으로 변한다.
    self.rear=0 #rear는 새로운 값을 추가 할때 self.rear=(self.rear+1)%self.Max_Qsize으로 변한다.

  def isEmpty(self):
    for x in self.Queue:
      if x != None:
        return False #return하면 반복중이더라도 함수가 끝난다.
    return True
  
  def enqueue(self,x):
    #원형큐의 포화상태: (self.rear+1)%self.Max_Qsize==self.front
    if self.isEmpty() == False and (self.rear+1)%self.Max_Qsize==self.front: #queue가 비어있지 않은데 rear==front가 되면 선입선출이 안되게 되서 오류가 발생
      print("Queue가 꽉 찼습니다.")
    else:
      self.rear=(self.rear+1)%self.Max_Qsize
      self.Queue[self.rear]=x

  #지우는 대신 None값으로 바꿔서 isEmpty가 정상작동될수 있게 만듬
  def dequeue(self):
    self.front=(self.front+1)%self.Max_Qsize
    R_value=self.Queue[self.front]
    self.Queue[self.front]=None
    return R_value

  def peek(self): #큐의 맨 앞에 있는 요소를 삭제하지 않고 반환
    return self.Queue[(self.front+1)%self.Max_Qsize] #front는 비어있는 값을 가리키므로 (self.front+1)%self.Max_Qsize로 맨앞값 가리키게 보정
  
  #예제코드에선 self.rear-self.front로 계산
  def size(self):
    count=0
    for x in self.Queue:
      if x != None:
        count+=1
    return count
  
  
  #지우는 대신 None값으로 바꿔서 isEmpty가 정상작동될수 있게 만듬
  def clear(self):
    for i in range(len(self.Queue)):
      self.Queue[i] = None
    #공백상태 조건: 전단(front)와 후단(rear)의 값이 동일
    self.front=0 #front값 0으로 초기화
    self.rear=0 #rear값 0으로 초기화
    
#원형큐의 장점: 삭제 삽입 전부 시간복잡도가 1로 계속 유지됨
#이유: 삽입하고 삭제할 위치를 항상 알수있게 됨
```

```python
A=Queue(10)
print(A.isEmpty())
A.enqueue(5)
print(A.isEmpty())
print(A.Queue)
print(A.dequeue())
print(A.Queue)
A.enqueue(6)
A.enqueue(7)
A.enqueue(8)
print(A.Queue)
print(A.peek())
print(A.size())
A.clear()
print(A.Queue)
print(A.isEmpty())
```

```python
B= Queue(5)
print(B.Queue)
for i in range(1,6):
  B.enqueue(i)
  print(B.Queue)
  print(B.rear)
  print(B.front)
print("------------")
for i in range(1,3):
  B.dequeue()
  print(B.Queue)
  print(B.rear)
  print(B.front)

for i in range(1,3):
  B.enqueue(i)
  print(B.Queue)
  print(B.rear)
  print(B.front)
```
