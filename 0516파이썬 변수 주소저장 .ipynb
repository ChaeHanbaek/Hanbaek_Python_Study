```python
a=3

print(id(a))
print(id(3)

#파이썬은 리터럴을 메모리에 저장, 그값을 저장하면 불러와서 참조함
#따라서 변수 a가 가리키는 주소와 3이 저장된 주소는 동일함
```

```python
a=[1,2,3]
b=[1,2,3]


print(id(a))
print(id(b))
#리터럴이 아닌 리스트등의 자료형은 내용물이 같더라도 주소가 다르다

c=[]
c.append(a) #append로 가져온 값은 원본과 주소가 동일
a.clear() #이후 a를 clear로 지워버리면 c에서도 가져온 주소의 목적지가 비어있다고 뜸
#a=[]꼴로 지우는거는 변수 a에 새로운 주소를 매핑해주는거기때문에 c에는 영향 없음


print(id(c[0]))
```
