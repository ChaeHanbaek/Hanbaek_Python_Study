```python
import random
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#디폴트값 설정
default_rand_nums = 10
default_min_of_rand_num = 0
default_max_of_rand_num = 10

str_inform = """이 프로그램은 임의의 숫자를 생성하고, 그 결과를  csv 파일로 저장, 통계 분석하는 기능을 제공합니다.\
\n숫자를 생성하기 위하여, 생성할 숫자의 갯수(n), 임의의 숫자 생성을 위한 범위의 최솟값(min), 최댓값(max)를 설정 할 수 있습니다.\
\n이 설정 옵션의 기본값은  n = {n}, min = {min}, max = {max}입니다.\
\n기본 옵션값으로 숫자를 생성하고자 하실 경우에는 'd'를 입력하고, 숫자 생성을 종료하실 경우에는 'q'를 입력하시면 됩니다.\n
""".format(n = default_rand_nums, min = default_min_of_rand_num, max = default_max_of_rand_num)

str_rand_num = "생성할 숫자의 옵션 n, min, max를 스페이스바로 구분하여 순서대로 입력해 주세요('q', 'd' 입력 가능) : "
```

```python
#_num, _min, _max를 요소로 받아 _min~_max 범위의 난수를 _num개 생성해 리스트에 넣어 그 리스트를 반환하는 메서드
def rand_num_gen(_nums = default_rand_nums, _min = default_min_of_rand_num, _max = default_max_of_rand_num): #디폴트값 설정
    res = []
    for i in range(_nums):
        num = random.randint(_min, _max)        
        res.append(num)
    return res
```

```python
#_data,(_tag=True)를 요소로 받아 추가로 파일이름을 입력받은 뒤 _data를 파일이름.csv로 저장하는 메서드
def save_csv(_data, _tag = True):
    file_name = input("csv 파일명을 입력해 주세요 : ");
    with open("{}.csv".format(file_name), mode = 'w', encoding = 'utf-8-sig') as f:
        try:
            w = csv.writer(f)
            #_tag가 True값이면 이런식으로 작동(중복불가한 태그들을 만듬)
            if _tag:
                #data의 길이를 측정
                max_len = 0
                for i in _data:
                    if len(i) > max_len:
                        max_len = len(i)
                
                #측정한 길이만큼 반복하여 set[index]자리에 _data 리스트의 각 요소를 넣음
                tags = []
                index = 0
                for i in range(max_len):
                    tags.append("data{}".format(index))
                    index += 1
                #그렇게 만든 태그 set을 한줄씩 csv에 기록
                w.writerow(tags)

            #_tag가 False이면 바로 한줄씩 csv에 기록
            for i in _data:
                w.writerow(i)
            print("{}.csv 파일 저장에 성공하였습니다.".format(file_name))            
            return True, file_name
            
        #에러 발생시 출력 문구
        except:
            print("csv 파일 저장에 실패하였습니다.")
            return False, None
```

```python
#요소없이 실행후 csv파일명을 입력받아 열어본뒤 list로 가공해 반환하는 메소드
def read_csv():
    while True:
        try:
            file_name = input("csv 파일명을 입력해 주세요 : ");
            with open("{}.csv".format(file_name), mode = 'r', encoding = 'utf-8-sig') as f:        
                #csv.reader사용해 한줄씩 읽은 리스트를 생성
                r = csv.reader(f)
                datas = []
                for data in r:
                    datas.append(data)
                
                print("csv 파일 불러오기에 성공하였습니다.")
                return datas
        #try except 예외 처리 되어있음
        except:
            print("{}.csv 파일이 존재하지 않습니다".format(file_name))     
```

```python
#_data, _mode(='r') (분기변수), _loc =[0] (인덱스역할)을 요소로 받아
#_mode='r'인경우 abs(_loc[0]) > abs(len(_data))일때만 _data[_loc[0]]을 반환
#_mode='c'인 경우 _data의 자료를 받아 abs(len(data)) > abs(len(_data))조건을 충족하는 data[_loc[0]]을 저장한 리스트를 반환
#_mode='p'인 경우 abs(_loc[0]) < abs(len(_data)) 와 abs(_loc[1]) < abs(len(_data[_loc[0]]))의 조건을 충족하면 _data[_loc[0]][_loc[1]]을 반환
#주어진 요소에 맞는 자료를 반환받는 메소드
def get_data(_data, _mode = 'r', _loc = [0]):

    if _mode == 'r':
        if abs(_loc[0]) > abs(len(_data)):
            print("데이터의 범위를 초과합니다.\n")
            return
        
        return _data[_loc[0]]

    elif _mode == 'c':
        res = []
        for data in _data:
            if abs(len(data)) > abs(_loc[0]):
                res.append(data[_loc[0]])
        
        if len(res) == 0:
            print("데이터의 범위를 초과합니다.\n")
            return

        return res        

    elif _mode == 'p':
        if abs(_loc[0]) < abs(len(_data)):
            if abs(_loc[1]) < abs(len(_data[_loc[0]])):
                return _data[_loc[0]][_loc[1]]
        
        print("데이터의 범위를 초과합니다.")

    #각 조건마다 if문으로 예외처리 되어있음
```

```python
#_data를 요소로 받아 평균(mean)을 구하는 메소드
def calc_mean(_data):
    int_data = list(map(int, _data))
    if len(_data) > 0:
        return sum(int_data)/len(int_data)
        
    #if, else로 데이터가 비었을때 대비한 예외처리 되어있음
    else:
        print("데이터가 존재하지 않습니다")
```

```python
#_data와 _mean을 요소로 받아 표준편차를 구하는 메소드 
def calc_var_std(_data, _mean):
    #_data 요소를 int형으로 변환
    int_data = list(map(int, _data))
    #_data길이가 0보다 크면 공식에 따라 분산과 표준편차를 계산후 분산과 표준편차를 반환
    if len(_data) > 0:
        var = 0
        for data in int_data:
            var += (data - _mean)**2
        var /= len(int_data)
        std = var**0.5
        return var, std
    # if, else로 데이터가 비었을때 대비한 예외처리 되어있음
    else:
        print("데이터가 존재하지 않습니다")
```

```python
#_data를 입력받아 누적분포를 구하고 누적분포 리스트 cdf=[[data요소][누적횟수]]와 그 발생빈도 prob=[[data요소][확률]]를 반환
def calc_cdf_prob(_data):
    #_data의 요소들을 int형으로 변환
    int_data = list(map(int, _data))
    #data길이가 0보다 크면 
    if len(_data) > 0:
        #elems라는 set을 통해 _data의 중복요소를 제거
        elems = set(int_data)
        #오름차순으로 정렬
        sort_elems = sorted(elems)
        #누적분포를 저장할 list와 그 와 연결된 cdf_index를 dictionary로 설정
        cdf = []
        cdf_index = {}
        index = 0
        #오름차순으로 정렬된 sort_elems요소들을 cdf리스트에 [elem, 0]식으로 집어넣고, cdf_index를 {str(elem:index(0부터 1씩커짐))}으로 집어넣음
        for elem in sort_elems:
            cdf.append([elem, 0])
            cdf_index.update({str(elem): index})
            index += 1
        # 정수형으로 변환한 int_data를 위에서 정리한 cdf_index에 키값으로 넣어서 cdf의 인덱스를 받고 cdf에 몇번씩 누적되는지를 기록
        for data in int_data:
            loc = cdf_index[str(data)] #인덱스 받기
            cdf[loc][1] += 1 #누적
        #len_data라는 변수를 int_data의 길이로 초기화
        len_data = len(int_data)
        #cdf을 이용해서 [data의 요소,그 발생빈도]꼴로 정리해 prob 리스트의 각 아이템으로 삼음
        prob = [[data[0], data[1]/len_data*100] for data in cdf]
        
        return cdf, prob
    else:
        print("데이터가 존재하지 않습니다")
```

```python
num_datas = []
#반복해서 값을 입력받아 난수 생성 리스트들을 만듬
while True:
    try:
        #입력받은 값들을 user_in() 리스트에 나눠서 넣음
        user_in = input(str_rand_num).split()
        #str_rand_num=n, min, max를 스페이스바로 구분하여 순서대로 입력해 주세요

        #q가 입력되면 종료
        if 'q' in user_in:
            print("숫자 생성을 종료합니다.\n")
            break
        
        #d를 입력하면 default로 정해둔 num=10, min=0, max=10으로 rand_num_gen()을 실행시켜 난수 리스트를 만들고 num_datas에 추가한뒤 출력
        elif 'd' in user_in:
            gen_nums = rand_num_gen()
            num_datas.append(gen_nums)
            print("생성된 데이터 : ", gen_nums)

        #num,min,max가 모두 입력 안되었을 때 예외처리
        elif len(user_in) != 3:
            print("입력 수가 잘못하였습니다. 다시 입력해주세요\n")    
        
        
        else:
            #입력받은 값을 int형으로 변환
            n = int(user_in[0])
            min = int(user_in[1])
            max = int(user_in[2])
            
            #min max값의 순서 보정
            if min > max:
                temp = max
                max = min
                min = temp
                print("min, max의 위치를 변경하였습니다. min : {}, max : {}".format(min, max))
            
            #n==0으로 입력됐을때 예외처리
            if n == 0:
                print("생성하고자 하는 숫자의 갯수가 0입니다. 다시 입력해주세요\n")
            
            #제대로 입력됐으면 rand_num_gen(n,min,max)로 난수 리스트를 만들고 num_datas에 저장후 출력
            gen_nums = rand_num_gen(_nums = n, _min = min, _max = max)
            num_datas.append(gen_nums)
            print("생성된 데이터 : ", gen_nums)    
    #입력 받은 것들이 정수형이나 q,d가 아닐경우의 예외처리
    except ValueError:
        print("입력을 잘못하였습니다. 다시 입력해주세요\n")

print("생성된 총 데이터 : \n", num_datas)
```

```python
#num_datas=[[][][],...,[]]을 save_csv의 요소로 넣어 작동(num_datas,true)
sucess, file_name = save_csv(num_datas)
#save_csv가 제대로 작동되면 True, file_name이 반환되고 오류나면 False, None이 반환됨
#제대로 작동되면 반환된 file_name을 요소로 삼아 read.csv()를 작동시켜 출력
if sucess:
    pd_data = pd.read_csv('{}.csv'.format(file_name), sep='\t')
    print(pd_data)
```

```python
#read_datas를 read_csv를 통해 불러온 파일의 데이터를 리스트로 저장
read_datas = read_csv()
#그 데이터를 출력
print("불러온 데이터 : \n", read_datas)
```

```python
#행 단위 데이터 처리
#인덱스=행의 번호가 되게 리스트 생성
file_row = []
#인덱스 0에 태그를 붙임
file_row.append(['data', 'mean', 'var', 'std', 'cdf', 'prob'])
#read_dats의 인덱스 1부터 반복(인덱스 0은 태그줄) 
for index in range(1, len(read_datas)):
    #mode=r로 get_data를 사용해서 행별로 데이터 추출
    row_data = get_data(read_datas, _mode = 'r', _loc = [index])
    #calc_mean으로 평균 계산
    mean = calc_mean(row_data)
    #분산, 표준편차 계산
    var, std = calc_var_std(row_data, mean)
    #누적분포함수, 빈도 계산
    cdf, prob = calc_cdf_prob(row_data)
    #결과값들 출력
    print("row data : ", row_data)
    print("mean : {:.2f}, variance : {:.2f}, std : {:.2f}".format(round(mean,3), round(var,3), round(std,3))) #{ : .2f}".format(실수 입력)소수점 2번째 자리까지만 출력, round(mean,3)소수점 3째자리에서 반올림
    print("cdf : ", cdf)
    print("prob : ", prob)
    #file_row 리스트에 [row_data, mean, var, std, cdf, prob]꼴로 저장
    file_row.append([row_data, mean, var, std, cdf, prob])

#save_csv로 성공여부변수, 파일이름 가져옴
sucess, file_name = save_csv(file_row, _tag = False) #첫줄 태그없이 사용
#저장 성공시 반환받았던 filename으로 데이터를 pd_data로 읽어오고 출력
if sucess:
    pd_data = pd.read_csv('{}.csv'.format(file_name), sep='\t')
    print(pd_data)
```

```python
#열 단위 데이터 처리
#가장 긴 열이 몇줄인지 저장해놓을 변수 max_len_col 설정
max_len_col = 0
for data in read_datas[1::]: #1부터 전체
    if len(data) > max_len_col:
        max_len_col = len(data)
    print(data)
#결과값을 저장할 file_col리스트에 인덱스 0 에 태그를 넣음
file_col = []
file_col.append(['data', 'mean', 'var', 'std', 'cdf', 'prob'])
for index in range(max_len_col):
    #mode='c'로 get_data사용해서 열데이터 추출
    col_data = get_data(read_datas[1::], _mode = 'c', _loc = [index])
    #열의 평균 계산
    mean = calc_mean(col_data)
    #분산, 표준편차 계산
    var, std = calc_var_std(col_data, mean)
    #누적분포 함수, 빈도 계산
    cdf, prob = calc_cdf_prob(col_data)
    #결과값들 출력
    print("col data : ", col_data)
    print("mean : {:.2f}, variance : {:.2f}, std : {:.2f}".format(round(mean,3), round(var,3), round(std,3)))
    print("cdf : ", cdf)
    print("prob : ", prob)
    #file_col에 [col_data, mean, var, std, cdf, prob]꼴로 넣는다.
    file_col.append([col_data, mean, var, std, cdf, prob])

#save_csv()를 태그없이 실행하고 성공결과, 파일이름을 받는다.
sucess, file_name = save_csv(file_col, _tag = False)
#성공시 read_csv()로 읽어온뒤 출력한다.
if sucess:
    pd_data = pd.read_csv('{}.csv'.format(file_name), sep='\t')
    print(pd_data)
```
