```python
def 랜덤함수(_a,_b,_k): #_a~_b범위에서 중복포함 랜덤으로 _k개 뽑아 리스트로 돌려주는 함수
  import random
  _list=[]
  while len(_list) < _k:
    _list.append(random.randint(_a,_b)) 
  return _list

def 줄별저장(_file,_list):
  i=0
  for x in _list:
    if i<(len(_list)-1):
      _file.write(str(x)+",")
      i+=1
    else:
      _file.write(str(x)+"\n")

#_lists=[[],[],[],[]] 꼴인 2차원 배열로 가정
def csv저장(_name,_lists):
  _f=open("{}.csv".format(_name), mode='w', newline="", encoding='utf-8-sig')
  for _row in _lists:
    줄별저장(_f,_row)
  _f.close

  #정수화 함수 #스트링으로 된값을 정수화시킴
def 정수화 (_row):
  for i in range(len(_row)):
    _row[i]=int(_row[i])
  return _row

#각행 평균
def 각행평균(_row):
  _sum=0
  for x in _row:
    x=int(x)
    _sum += x
  _average = _sum/len(_row)
  return _average

#각행 편차 : 편차는 리스트로 만들어두지만 최종 리턴값은 표준편차
def 각행편차(_row):
  _average=각행평균(_row)
  편차리스트=[]
  for x in _row:
    x=int(x)
    _deviation = x-_average
    편차리스트.append(_deviation)
  #print(편차리스트) #정상작동 확인용

  _dsum=0
  for y in 편차리스트:
    _dsum += y*y
  _variance = _dsum/len(_row)
  _sd = _variance **0.5
  return _sd
#계산기로 도출한 값과 같음

#행별누적분포, 각 요소가 몇번 반복됐는지 누적해서 저장
#누적분포는 dictionary로 저장해 정수값:누적분포로 관리하는게 이상적일듯
def 행별누적분포(_row):
  _행별누적분포dic=dict.fromkeys(_row)
  for x in _row:
    _행별누적분포dic[x]=_row.count(x)
  return(_행별누적분포dic)
```

```python
#저장된 파일 읽어서 처리해 리스트로 저장
import csv
f=open("{}.csv".format(filename),mode='r',encoding='utf-8-sig')
rd=csv.reader(f)
전체정보리스트=[]
for row in rd:
  정수화(row)
  행별정보리스트=[각행평균(row),각행편차(row),행별누적분포(row)]
  전체정보리스트.append(행별정보리스트)

#print(전체정보리스트) #정상작동확인용
f.close()


#전체정보리스트[i][2] 는 i번째 열의 누적분포
datakeylist=[] #누적분포dic에서 key값을 모아낼 리스트
for i in range(len(전체정보리스트)):
  _tempkeys=전체정보리스트[i][2].keys() #각행 누적분포에서 키값만 뽑아냄
  datakeylist.extend(_tempkeys) #일단 통째로 넣음

datakeylist = set(datakeylist)
datakeylist = list(datakeylist) #set으로 바꿨다 list로 다시바꾸는거를 통해 편하게 중복제거
datakeylist.sort()

#print(datakeylist) #정상작동 확인용

#datakeylist바탕으로 전체 누적분포 구하기
#dic=dict.fromkeys(seq[,v]) : 시퀀스의 값을 키로 갖고 그 값을 v로 갖는 새로운 dictionary 생성 (v는 보통 생략됨 생략하면 None값 가짐)
전체누적분포=dict.fromkeys(datakeylist,0) #초기값 0
t_sum=0 #발생확률 구하기위해 필요
for x in datakeylist:
  d_sum=0
  for i in range(len(전체정보리스트)):
    _t=전체정보리스트[i][2]
    if _t.get(x)!=None:
      d_sum += 전체정보리스트[i][2][x]
      t_sum += 전체정보리스트[i][2][x]
  전체누적분포[x]=d_sum

#print(전체누적분포) #정상작동확인용
```

```python
#가공한 리스트로 형식에 맞춰 새로운 파일 저장
print(전체정보리스트)
#위에서 구한 전체정보리스트와 datakeylist 바탕 각행별 누적분포 파일로 저장
f=open("결과물.csv", mode='w',encoding='utf-8-sig')
#첫번째 행에 전체정보리스트에 있는 데이터들과 평균, 표준편차 라벨 입력
f.write("구분,")
for x in datakeylist:
  f.write("data: "+str(x)+',')
f.write("평균,표준편차\n")

#각 행별로 데이터 입력
for i in range(len(전체정보리스트)):
  f.write("{}행,".format(i+1))
  for x in datakeylist:
    #dic.get(key[,d])
    _t=전체정보리스트[i][2]
    if _t.get(x)!=None:
      f.write(str(_t[x])+',')
    else:
      f.write(str(0)+',')
  f.write(str(전체정보리스트[i][0])+','+str(전체정보리스트[i][1])+'\n')

#마지막에 전체누적분포와 발생확률 입력
#전체 누적분포
f.write("전체 누적분포,")
for x in datakeylist:
  f.write(str(전체누적분포[x])+',')
f.write("총 시행수,{}".format(str(t_sum)))
f.write('\n')
#발생확률
f.write("발생확률,")
for x in datakeylist:
  f.write(str(전체누적분포[x]/t_sum)+',')
f.write('\n')
```
