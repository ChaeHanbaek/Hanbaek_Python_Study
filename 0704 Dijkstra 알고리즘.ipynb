```python
#주어진 거리가 최단거리인지 확인해보는 함수
INF = 9999
def choose_vertex(dist,found): #거리와 최단거리 찾은지 확인하는 리스트를 받아
  min = INF
  minpos = -1
  for i in range(len(dist)):
    if dist[i]<min and found[i]==False: #찾지않았을때(false) 최단거리를 찾아서
      min = dist[i]
      minpos = i
  return minpos; #그 최단거리의 index를 넘겨줌
```

```python
#vertex 노드명 순서대로 넣은 행렬
#adj 노드간 가중치 순서대로 들어간 이중행렬
#start 시작노드의 vertex에서의 index
def shortest_path_dijkstra(vtx,adj,start): 
  vsize=len(vtx)
  dist= list(adj[start]) #시작 노드와 다른 노드 사이의 거리를 가중치 행렬에서 가져옴
  path = [start] * vsize #path list를 노드수만큼 만들어둠 기본값 start
  found = [False] * vsize #found list를 노드수만큼 만들어둠 기본값 False//찾았으면 True, 못찾았으면 False
  found[start] = True #found list의 start index만 True로 전환
  dist[start] = 0 #자기 자신과의 거리는 0으로 설정
  for i in range(vsize):
    print("Step%2d: "%(i+1),dist) #각 반복에서의 시작점과 다른 노드사이 최단거리 구한거 출력
    u=choose_vertex(dist,found) #최단거리 노드의 index를 받아옴
    found[u] = True #그 index에 해당하는걸 found에서 True로 전환
    for w in range(vsize): #노드수만큼 반복
      if not found[w]: #found되지 않은 노드에서 (not False ==True)
        if dist[u] + adj[u][w] < dist[w]: #지금 찾은 최단거리+거기서 그노드로 가는 거리가 최단거리보다 짧은지 비교해
          dist[w] = dist[u] + adj[u][w] #만약 맞으면 dist에 들어있는 최단거리 값을 바꾼다.
          path[w] = u #거길 지나는 path값도 start가 아니라 지금 확인하고 있는 노드로 변환
  return path
```

```python
#사전자료
I = float("inf") #무한대값을 가진 변수 설정
#         A  B  C  D  E  F  G
weight=[
        [ 0, 7, I, I, 3,10, I],
        [ 7, 0, 4,10, 2, 6, I],
        [ I, 4, 0, 2, I, I, I],
        [ I,10, 2, 0,11, 9, 4],
        [ 3, 2, I,11, 0, I, 5],
        [10, 6, I, 9, I, 0, I],
        [ I, I, I, 4, 5, I, 0]
        ]
vertex = ['A','B','C','D','E','F','G']
#구동부
print("Shortest Path By dijkstra Algorithm")
start = 0
path = shortest_path_dijkstra(vertex,weight,start) #위의 함수를 돌려 경로를 받아옴
for end in range(len(vertex)):
  if end != start:
    # "%s" % () : %s(문자열), %d(정수), %f(실수)타입과 % ()안에 들어간 변수가 일치하면 순서대로 출력
    print("[최단경로: %s->%s] %s" %
          (vertex[start],vertex[end],vertex[end]),end='')
    while (path[end] != start): #path에 들어간 경로들을 순서대로 뱉어냄
      print(" <- %s" %vertex[path[end]],end='')
      end=path[end]
    print(" <=%s" % vertex[path[end]])
```
