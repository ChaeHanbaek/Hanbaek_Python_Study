```python
#노드 클래스
class node:
  def __init__(self,elem,link=None):
    self.data=elem
    self.link=link
```

```python
#연결된 리스트
class LinkedList:
  def __init__(self):
    self.head=None

  def insert(self,pos,e):
    Link=self.head      
    if self.head==None: #첫부분이 비어있으면 다르게 처리해줘야됨
      HEAD=node(None,None)
      self.head=HEAD
      print(self.head)
      Link=HEAD
    for i in range(pos-1): #pos번째 값에 접근하기 위해 pos-1만큼 반복(1번째는 head값이므로)
      print(Link)
      if Link==None or Link.link==None: #비어있으면 빈 노드를 만들어서 넘어감     
        New=node(None,None)
        Link.link=New
        Link=New
      else:
        Link=Link.link
    print("{}번째 노드에 입력: {}".format(pos,e))
    Link.data=e # 위 반복문을 통해 원하던 칸에 접근했으니 거기있는 데이터값을 입력받은 값으로 바꿈
      
```

```python
#테스트
LL=LinkedList()
LL.insert(1,1)
LL.insert(2,2)

LL.insert(7,7)
LL.insert(8,8)
print(LL.head.link.data)
print(LL.head.link.link.data)
print(LL.head.link.link.link.link.link.link.data)
```
