```python
#노드 클래스
class node:
  def __init__(self,elem,link=None):
    self.data=elem
    self.link=link
```

```python
#연결된 스택
class LinkedStack:
  def __init__(self):
    self.top=None #첫번째 스택의 주소를 할당

  def isEmpty(self):
    return self.top==None

  def push(self,e):
    #기존 스택,빈스택 구별 필요 없음
    New=node(e,link=self.top) #맨위에 추가할때 링크에는 이전 맨 위에 있던 스택의 주소를 할당해야함
    self.top=New #top이 가진 주소를 새로 만든거로 바꿈

  def pop(self):
    if self.isEmpty():
      print("비었습니다.")
    else:
      poped=self.top #가장 위의 스택을 불러와서
      poped_data=poped.data #그값을 변수에 할당
      self.top=poped.link #top의 주소를 그 스택의 link값으로 바꾸고
      return poped_data #변수로 할당한 값을 반환
  
  def peek(self):
    if self.isEmpty():
      print("비었습니다.")
    else:
      return self.top.data #top에 연결된 스택의 값을 반환

  def clear(self):
    self.top=None #top에 할당된 주소값을 날리면 공백상태

  def size(self):
    AL=self.top
    if self.isEmpty():
      return 0
    else:
      stackcount=0
      while True:
        stackcount=stackcount+1
        AL=AL.link
        if AL==None:
          break
      return stackcount
  
  def display(self,msg="LinkedStack: "):
    print(msg, end=' ')
    node=self.top
    while not node == None:
      print(node.data, end=' ')
      node= node.link
    print()


```

```python
A=LinkedStack()
print(A.isEmpty())
A.push(5)
print(A.peek())
A.push(6)
print(A.peek())
A.push(7)
print(A.peek())
A.push(8)
print(A.peek())
print(A.size())
print(A.isEmpty())
print("---------")
print(A.pop())
print(A.pop())
print(A.pop())
print(A.pop())
print(A.size())
print(A.isEmpty())
print("---------")
A.push(5)
A.push(5)
A.push(5)
A.push(5)
A.push(5)
A.clear()
print(A.size())
print(A.isEmpty())
```
