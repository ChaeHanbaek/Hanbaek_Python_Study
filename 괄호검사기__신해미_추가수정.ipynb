```python
class Stack:

  def __init__(self):
    self.stack=[]

  def isEmpty(self):
    if len(self.stack) == 0:
      print("True")
      return True #리턴값 빼먹음
    else:
      print("False")
      return False #리턴값 빼먹음

  def push(self, e):
    self.stack.append(e)
    return self.stack

  def pop(self):
     a=self.stack.pop() #이부분 잘못 변수 넘겨주는걸 self.stack 으로 되어있음
     return a #이렇게 변수할당해주고 그 변수를 넘겨줘야 정상

  def peek(self):
    print(self.stack)

  def size(self):
    stack_len=len(self.stack)
    print(stack_len)

  def clear(self):
    self.stack.clear()
    return self.stack
```

```python
def check(Str):
  s=Stack()  
  for i in Str:
    if i in '[{(':
      s.push(i)
      print(s.stack)
    elif i in ']})':
      if s.isEmpty():
        return (False, "괄호의 시작이 부족하거나, 괄호의 끝이 더 많습니다.")
      else:       
        j = s.pop()
        print("확인:{}와{}".format(j,i))
        if (i=="]" and j !="[") or (i=="}" and j !="{") or (i==")" and j!="("):
          print("{}에 대응하는 괄호가 없습니다.".format(j))
          return (False, "{}에 대응하는 괄호가 없습니다.".format(j)) #i가 아니라 j
  print('-------------')
  print(s.stack)
  if s.isEmpty():
    return (True, "모든 괄호가 적절하게 대응합니다")
  else:
    return(False, "대응되지 못한 괄호가 남아 있습니다")

st = ("{A[(i+1)]=0}", "if((i==0)&&(j==0)", "A[(i+1])=0;")
for a in st:
  print('-------------')
  b,c= check(a)
  print(a,":", b,c)
  print('-------------')
```
