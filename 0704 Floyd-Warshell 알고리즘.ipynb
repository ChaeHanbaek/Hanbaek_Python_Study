```python
#사전자료
I = 9999 #무한대값을 가진 변수 설정
#         A  B  C  D  E  F  G
weight=[
        [ 0, 7, I, I, 3,10, I],
        [ 7, 0, 4,10, 2, 6, I],
        [ I, 4, 0, 2, I, I, I],
        [ I,10, 2, 0,11, 9, 4],
        [ 3, 2, I,11, 0, I, 5],



        
        [10, 6, I, 9, I, 0, I],
        [ I, I, I, 4, 5, I, 0]
        ]
vertex = ['A','B','C','D','E','F','G']
```

```python
INF = 9999
def printA(A):
  vsize=len(A)
  print("==============================")
  for i in range(vsize):
    for j in range(vsize):
      if (A[i][j]==INF):
        print(" INF ",end='')
      else:
        print("%4d "%A[i][j],end='')
      print("");
```

```python
def shortest_path_floyd(vertex,adj):
  vsize = len(vertex)
  A = list(adj)
  for i in range(vsize):
    A[i] = list(adj[i])
  
  #삼중for문으로 한번에 최단경로 탐색
  for k in range(vsize):
    for i in range(vsize):
      for j in range(vsize):
        if (A[i][k] +A[k][j] < A[i][j]):
            A[i][j] = A[i][k] + A[k][j]
    printA(A)
```

```python
shortest_path_floyd(vertex,weight)
```
