```python
#학년별 데이터 베이스 부분
#1학년
국어_G1_C1=[None,0,0,0,0,0]
국어_G1_C2=[None,0,0,0,0]
국어_G1_C3=[None,0,0,0,0,0]
국어_G1_C4=[None,0,0,0]
국어_G1_C5=[None,0,0,0,0]
#2학년
국어_G2_C1=[None,0,0,0,0,0]
국어_G2_C2=[None,0,0,0,0]
국어_G2_C3=[None,0,0,0,0,0]
국어_G2_C4=[None,0,0,0]
국어_G2_C5=[None,0,0,0,0]
#3학년
국어_G3_C1=[None,0,0,0,0,0]
국어_G3_C2=[None,0,0,0,0]
국어_G3_C3=[None,0,0,0,0,0]
국어_G3_C4=[None,0,0,0]
국어_G3_C5=[None,0,0,0,0]
#4학년
국어_G4_C1=[None,0,0,0,0,0]
국어_G4_C2=[None,0,0,0,0]
국어_G4_C3=[None,0,0,0,0,0]
국어_G4_C4=[None,0,0,0]
국어_G4_C5=[None,0,0,0,0]
#5학년
국어_G5_C1=[None,0,0,0,0,0]
국어_G5_C2=[None,0,0,0,0]
국어_G5_C3=[None,0,0,0,0,0]
국어_G5_C4=[None,0,0,0]
국어_G5_C5=[None,0,0,0,0]
#6학년
국어_G6_C1=[None,0,0,0,0,0]
국어_G6_C2=[None,0,0,0,0]
국어_G6_C3=[None,0,0,0,0,0]
국어_G6_C4=[None,0,0,0]
국어_G6_C5=[None,0,0,0,0]

#학년 구성
국어_G1=[None,국어_G1_C1,국어_G1_C2,국어_G1_C3,국어_G1_C4,국어_G1_C5]
국어_G2=[None,국어_G2_C1,국어_G2_C2,국어_G2_C3,국어_G2_C4,국어_G2_C5]
국어_G3=[None,국어_G3_C1,국어_G3_C2,국어_G3_C3,국어_G3_C4,국어_G3_C5]
국어_G4=[None,국어_G4_C1,국어_G4_C2,국어_G4_C3,국어_G4_C4,국어_G4_C5]
국어_G5=[None,국어_G5_C1,국어_G5_C2,국어_G5_C3,국어_G5_C4,국어_G5_C5]
국어_G6=[None,국어_G6_C1,국어_G6_C2,국어_G6_C3,국어_G6_C4,국어_G6_C5]
#학교 구성
국어=[None,국어_G1,국어_G2,국어_G3,국어_G4,국어_G5,국어_G6]


#모든 index 0위치에 None값을 넣어 보다 직관적이게 자료형에 접근 가능하게 바꿈 e.g) 1학년 1반 1번 학생에 접근할때 국어[1][1][1]로 참조 가능

#학생들의 국어점수 입력 및 수정 함수
#점수리스트는 index 0에 None값을 넣은 꼴로 만들었다고 가정 e.g) [None,50,60,70,80,80]
#학년별로 입력
def 학년별입력(학년,학년별점수리스트):
  국어[학년]=학년별점수리스트
#반별로 입력
def 반별입력(학년,반,반별점수리스트):
  국어[학년][반]=반별점수리스트

#학생별 입력 번호는 1부터 시작
def 학생별입력(학년,반,번호,학생점수):
  국어[학년][반][번호]=학생점수

#점수 검증 함수
def 오류찾기(국어):
  오류학생리스트=[] #오류가 있는 학생의 [학년,반,번호]데이터를 입력한 리스트를 만들어서 반환
  for i in range(1,len(국어)): #학년 인덱스 #0을 더미로 썼으니 1부터 시작
    for j in range(1,len(국어[i])): #반 인덱스 #0을 더미로 썼으니 1부터 시작
      for k in range(1,len(국어[i][j])): #학생 인덱스 #0을 더미로 썼으니 1부터 시작
        if 국어[i][j][k] > 100 or 국어[i][j][k] <0:
          #print("{}학년 {}반 {}번 학생의 성적에 문제가 있습니다.".format(i,j,k))
          오류학생리스트.append([i,j,k])
  return 오류학생리스트


#각학년에서 반별 평균 계산, 
def 학년별반평균(학년):
  반평균=[None] #인덱스가 곧 반을 나타나게 인덱스0을 None값 넣어둠
  for j in range(1,len(국어[학년])): #반 인덱스 0을 더미로 썼으니 1부터 시작
    sum = 0
    for k in range(1,len(국어[학년][j])): #학생 인덱스 0을 더미로 썼으니 1부터 시작
      sum = sum + 국어[학년][j][k]
    반평균.append(sum/(len(국어[학년][j])-1)) #인덱스 0은 더미니 학생수는 len()-1
  return 반평균

#가장 높은 반과, 낮은 반을 각각 보고
def 학년별최고최하(학년):
  반평균=학년별반평균(학년)
  max=반평균[1]
  for x in range(1,len(반평균)):
    if 반평균[x]>max:
      max = 반평균[x]
  min = 반평균[1]
  for x in range(1,len(반평균)):
    if 반평균[x]<min:
      min = 반평균[x]
  return max, min


#학년의 평균 점수 보다 낮은 점수를 가진 학생들은 방과 후에 보충 수업을 해야함, 그래서 그인원이 누구인지 찾아내야함, 명단 작성 요함
#학년평균도 [None,1학년평균,2학년평균,3학년평균,4학년평균,5학년평균,6학년평균]으로 만들어야함
def 학년별평균(국어): #학년별 반평균이용해 학년평균 구하기
  학년별평균목록=[None] #인덱스가 곧 학년을 나타나게 인덱스0dp None값 넣어둠 
  for i in range(1,len(국어)): #학년 인덱스 i, 0을 더미로 썼으니 1부터 시작
    sum = 0
    x = 학년별반평균(i)
    for j in range(1,len(학년별반평균(i))): #학년별반평균(i)의 값을 인덱스 1에서부터 더한다.
     sum = sum + x[j]
    sum = sum / (len(학년별반평균(i))-1) #index0을 제외한 크기로 나눈다
    학년별평균목록.append(sum)
  return 학년별평균목록
      

def 보충수업(국어):
  보충수업리스트=[] #보충수업해야되는 학생의 [학년,반,번호]데이터를 입력한 리스트를 만들어서 반환
  학년평균목록=학년별평균(국어)
  for i in range(1,len(국어)): #학년 인덱스 #0을 더미로 썼으니 1부터 시작
    for j in range(1,len(국어[i])): #반 인덱스 #0을 더미로 썼으니 1부터 시작
      for k in range(1,len(국어[i][j])): #학생 인덱스 #0을 더미로 썼으니 1부터 시작
        if 국어[i][j][k] < 학년평균목록[i]:
          #print("{}학년 {}반 {}번 학생은 보충수업해야 합니다.".format(i,j,k))
          보충수업리스트.append([i,j,k])
  return 보충수업리스트

학년1 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,50,40,20],[None,70,80,60],[None,50,40,50,30]]
학년2 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,50,40,20],[None,70,80,60],[None,50,40,50,30]]
학년3 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,50,40,20],[None,70,80,60],[None,50,40,50,30]]
학년4 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,-50,-40,20],[None,70,80,60],[None,50,40,50,30]]
학년5 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,50,40,20],[None,70,80,60],[None,50,40,50,30]]
학년6 = [None,[None,70,50,60,70,80],[None,70,50,60,120],[None,100,90,50,40,20],[None,70,80,60],[None,50,40,50,30]]

학년별입력(1,학년1)
학년별입력(2,학년2)
학년별입력(3,학년3)
학년별입력(4,학년4)
학년별입력(5,학년5)
학년별입력(6,학년6)
print(국어)
print(오류찾기(국어))
print("------수정--------")
학생별입력(1,2,4,50)
print(오류찾기(국어))
print(학년별반평균(1))
print(학년별최고최하(1))
print(학년별평균(국어))
print(보충수업(국어))
```

```python
"""교수님 조언 리스트[][][]꼴로 하면 한번에 사용하는 메모리가 많아지기 때문에
  for i in range(1,len(국어)): 
   t_in = 국어[i]
    for j in range(1,len(t_in )): 식으로 반복문에서 리스트를 조금씩 끊어서 사용하는 임시 변수로 바꿔 사용하는게 좋다.
"""
#위 조언 반영해서 함수들 개선
#학년별 데이터 베이스 부분
#1학년
국어_G1_C1=[None,0,0,0,0,0]
국어_G1_C2=[None,0,0,0,0]
국어_G1_C3=[None,0,0,0,0,0]
국어_G1_C4=[None,0,0,0]
국어_G1_C5=[None,0,0,0,0]
#2학년
국어_G2_C1=[None,0,0,0,0,0]
국어_G2_C2=[None,0,0,0,0]
국어_G2_C3=[None,0,0,0,0,0]
국어_G2_C4=[None,0,0,0]
국어_G2_C5=[None,0,0,0,0]
#3학년
국어_G3_C1=[None,0,0,0,0,0]
국어_G3_C2=[None,0,0,0,0]
국어_G3_C3=[None,0,0,0,0,0]
국어_G3_C4=[None,0,0,0]
국어_G3_C5=[None,0,0,0,0]
#4학년
국어_G4_C1=[None,0,0,0,0,0]
국어_G4_C2=[None,0,0,0,0]
국어_G4_C3=[None,0,0,0,0,0]
국어_G4_C4=[None,0,0,0]
국어_G4_C5=[None,0,0,0,0]
#5학년
국어_G5_C1=[None,0,0,0,0,0]
국어_G5_C2=[None,0,0,0,0]
국어_G5_C3=[None,0,0,0,0,0]
국어_G5_C4=[None,0,0,0]
국어_G5_C5=[None,0,0,0,0]
#6학년
국어_G6_C1=[None,0,0,0,0,0]
국어_G6_C2=[None,0,0,0,0]
국어_G6_C3=[None,0,0,0,0,0]
국어_G6_C4=[None,0,0,0]
국어_G6_C5=[None,0,0,0,0]

#학년 구성
국어_G1=[None,국어_G1_C1,국어_G1_C2,국어_G1_C3,국어_G1_C4,국어_G1_C5]
국어_G2=[None,국어_G2_C1,국어_G2_C2,국어_G2_C3,국어_G2_C4,국어_G2_C5]
국어_G3=[None,국어_G3_C1,국어_G3_C2,국어_G3_C3,국어_G3_C4,국어_G3_C5]
국어_G4=[None,국어_G4_C1,국어_G4_C2,국어_G4_C3,국어_G4_C4,국어_G4_C5]
국어_G5=[None,국어_G5_C1,국어_G5_C2,국어_G5_C3,국어_G5_C4,국어_G5_C5]
국어_G6=[None,국어_G6_C1,국어_G6_C2,국어_G6_C3,국어_G6_C4,국어_G6_C5]
#학교 구성
국어=[None,국어_G1,국어_G2,국어_G3,국어_G4,국어_G5,국어_G6]


#모든 index 0위치에 None값을 넣어 보다 직관적이게 자료형에 접근 가능하게 바꿈 e.g) 1학년 1반 1번 학생에 접근할때 국어[1][1][1]로 참조 가능

#학생들의 국어점수 입력 및 수정 함수
#점수리스트는 index 0에 None값을 넣은 꼴로 만들었다고 가정 e.g) [None,50,60,70,80,80]
#학년별로 입력
def 학년별입력(학년,학년별점수리스트):
  국어[학년]=학년별점수리스트
#반별로 입력
def 반별입력(학년,반,반별점수리스트):
  국어[학년][반]=반별점수리스트

#학생별 입력 번호는 1부터 시작
def 학생별입력(학년,반,번호,학생점수):
  국어[학년][반][번호]=학생점수

#점수 검증 함수
def 오류찾기(국어):
  오류학생리스트=[] #오류가 있는 학생의 [학년,반,번호]데이터를 입력한 리스트를 만들어서 반환
  for i in range(1,len(국어)): #학년 인덱스 0을 더미로 썼으니 1부터 시작
    x_in = 국어[i]
    for j in range(1,len(x_in)): #반 인덱스 0을 더미로 썼으니 1부터 시작
      y_in =x_in[j]
      for k in range(1,len(y_in)): #학생 인덱스 0을 더미로 썼으니 1부터 시작
        if y_in[k] > 100 or y_in[k] <0:
          #print("{}학년 {}반 {}번 학생의 성적에 문제가 있습니다.".format(i,j,k))
          오류학생리스트.append([i,j,k])
  return 오류학생리스트

#각학년에서 반별 평균 계산, 
def 학년별반평균(학년):
  반평균=[None] #인덱스가 곧 반을 나타나게 인덱스0을 None값 넣어둠
  x_in = 국어[학년]
  for j in range(1,len(x_in)): #반 인덱스 0을 더미로 썼으니 1부터 시작
    sum = 0
    y_in = x_in[j]
    for k in range(1,len(y_in)): #학생 인덱스 0을 더미로 썼으니 1부터 시작
      sum = sum + y_in[k]
    반평균.append(sum/(len(y_in)-1)) #인덱스 0은 더미니 학생수는 len()-1
  return 반평균

#가장 높은 반과, 낮은 반을 각각 보고
def 학년별최고최하(학년):
  반평균=학년별반평균(학년)
  max = 반평균[1]
  min = 반평균[1]
  for x in range(1,len(반평균)):
    if 반평균[x]>max:
      max = 반평균[x]
    if 반평균[x]<min:
      min = 반평균[x]
  return max, min


#학년의 평균 점수 보다 낮은 점수를 가진 학생들은 방과 후에 보충 수업을 해야함, 그래서 그인원이 누구인지 찾아내야함, 명단 작성 요함
#학년평균도 [None,1학년평균,2학년평균,3학년평균,4학년평균,5학년평균,6학년평균]으로 만들어야함
def 학년별평균(국어): #학년별 반평균이용해 학년평균 구하기
  학년별평균목록=[None] #인덱스가 곧 학년을 나타나게 인덱스0dp None값 넣어둠 
  for i in range(1,len(국어)): #학년 인덱스 i, 0을 더미로 썼으니 1부터 시작
    sum = 0
    x = 학년별반평균(i)
    for j in range(1,len(학년별반평균(i))): #학년별반평균(i)의 값을 인덱스 1에서부터 더한다.
     sum = sum + x[j]
    sum = sum / (len(학년별반평균(i))-1) #index0을 제외한 크기로 나눈다
    학년별평균목록.append(sum)
  return 학년별평균목록
      

def 보충수업(국어):
  보충수업리스트=[] #보충수업해야되는 학생의 [학년,반,번호]데이터를 입력한 리스트를 만들어서 반환
  학년평균목록=학년별평균(국어)
  for i in range(1,len(국어)): #학년 인덱스 0을 더미로 썼으니 1부터 시작
    x_in = 국어[i]
    for j in range(1,len(x_in)): #반 인덱스 0을 더미로 썼으니 1부터 시작
      y_in = x_in[j]
      for k in range(1,len(y_in)): #학생 인덱스 0을 더미로 썼으니 1부터 시작
        if y_in[k] < 학년평균목록[i]:
          #print("{}학년 {}반 {}번 학생은 보충수업해야 합니다.".format(i,j,k))
          보충수업리스트.append([i,j,k])
  return 보충수업리스트

#임의의 난수를 받는 식으로 리스트 개조
import random

학년1 = [None,[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]
학년2 = [None,[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]
학년3 = [None,[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]
학년4 = [None,[None,random.randrange(1,101),-50,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]
학년5 = [None,[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]
학년6 = [None,[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)],[None,random.randrange(1,101),random.randrange(1,101),random.randrange(1,101),random.randrange(1,101)]]

학년별입력(1,학년1)
학년별입력(2,학년2)
학년별입력(3,학년3)
학년별입력(4,학년4)
학년별입력(5,학년5)
학년별입력(6,학년6)
print(국어)
print(오류찾기(국어))
print("------수정--------")
학생별입력(4,1,2,50)
print(오류찾기(국어))
print(학년별반평균(4))
print(학년별최고최하(4))
print(학년별평균(국어))
print(보충수업(국어))
```
