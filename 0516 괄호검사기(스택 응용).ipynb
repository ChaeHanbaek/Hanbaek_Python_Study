```python
class STACK():
  def __init__(self):
    self.lst=[]
  
  def isEmpty(self):
    if len(self.lst)==0:
      return True
    else:
      return False
  
  def push(self,e):
    self.lst.append(e)
    #print("STACK 맨위에 {}을 추가하였습니다.".format(e))
  
  def pop(self):
    t_v=self.lst.pop()
    #print("{}을 STACK 맨 위에서 꺼냈습니다.".format(t_v))
    return t_v
    

  def peek(self):
    #print("{}은 STACK 맨 위에 있는 요소입니다.".format(self.lst[(len(self.lst)-1)]))
    return self.lst[len(self.lst)-1]

  def size(self):
    t_count=0
    for x in self.lst:
      t_count +=1
    return t_count

  def clear(self):
    self.lst=[]

  def display(self):
    print(self.lst)
```

```python
#{{( 없는데 )]}로 시작하면 경고
#Last in First out으로 마지막에 들어온 {{(에 맞는 )]}가 와야함
#LIFO=clst[len(clst)-1]로 설정해서 그에 맞게 작동하게 하면 될듯 dictionary에 check={"{":"}","[":"]","(":")"}이렇게 넣고 check[LIFO]로 매칭시키면 될듯

def bracket_checker(str):
  clst=STACK()
  check={"{":"}","[":"]","(":")"}
  count=0
  LIFO=0
  err=0

  for c in str:
    count += 1
    #print("{}번째".format(count)) #작동확인용
    #print(clst.lst)
    #print(str[:count])
    if len(clst.lst)>0:
      LIFO=clst.lst[len(clst.lst)-1]
      
    elif len(clst.lst)==0:
      LIFO=None
    #print(LIFO)

    if (c=='{' or c=='[' or c=='('): # c in '{[('로 하나의 판독문으로 쓸 수 있음(스트링에 속한 요소들)
      clst.push(c)

    elif (c=='}' or c==']' or c==')'):
      if len(clst.lst)==0: 
        print("'{}'".format(str[:count]))
        print("{}'가 잘못위치 하고 있습니다.".format(c))
        err+=1

      elif check[LIFO]!=c:
        print("'{}'".format(str[:count]))
        print("{}가 {}와 짝이 맞지않습니다.".format(LIFO,c))
        clst.pop() #예시문에선 에러없음 #좀길어지면 불안하긴한데 큰이상은 없는듯
        print(clst.lst)
        err+=1

      elif check[LIFO]==c:
        clst.pop()

  if len(clst.lst)>0:
    print("{}가 남았습니다.".format(clst.lst))
  elif len(clst.lst)==0 and err==0:
    print("괄호에 이상이 없습니다.")
```
